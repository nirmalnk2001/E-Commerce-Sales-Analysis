```python
import pyodbc
import pymysql
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel("Superstore_USA.xlsx")
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#filled mean value in the missing value
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

## Order Priority

```python
df['Order Priority'].value_counts()
```

```python
#two critical values are there in order priority
df['Order Priority'].unique()
```

```python
df['Order Priority']=df['Order Priority'].replace("Critical ","Critical")
```

```python
df['Order Priority'].value_counts()
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x="Order Priority",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count No: of Order Priority")
plt.savefig("Count Of Order Priority.jpg")
plt.show()
```

## Ship Mode 

```python
df['Ship Mode'].value_counts()
```

```python
x=df['Ship Mode'].value_counts().index
y=df['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
plt.show()
```

```python
plt.figure(figsize=(5,4))
ax=sns.countplot(x="Ship Mode",data=df,color="Blue")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Ship Modes")
plt.show()
```

```python
plt.figure(figsize=(5,4))
ax=sns.countplot(x="Ship Mode",data=df,hue="Product Category")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("ShipModes_ProdCat.jpg")
plt.show()
```
